```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
#contar a frequência de cada tipo de negócio
tipo_counts=df['Tipo de negócio'].value_counts().reset_index()
tipo_counts.columns=['Tipo de negócio','Quantidade']
```

🔹 2. Distribuição por bairro

Objetivo: Mapear concentração geográfica dos negócios.

Valor: Descobrir onde estão os polos mais fortes e onde há menos concorrência (oportunidade de penetração).

```python
bairro_counts = df['Bairro'].value_counts().reset_index()
bairro_counts.columns = ['Bairro', 'Quantidade']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=bairro_counts, x='Quantidade', y='Bairro', palette='viridis')
plt.title('Distribuição por Bairro')
plt.xlabel('Quantidade de Negócios')
plt.ylabel('Bairro')
plt.tight_layout()
plt.show()
```

🔹 3. Situação de regularização dos negócios

Objetivo: Ver quantos negócios estão regularizados e quais não.

Valor: Negócios não regularizados são potenciais clientes imediatos para consultoria em boas práticas, licenciamento e adequação sanitária.

```python
#explorar o conteúdo da coluna de Regularizado
df['Regularizado'].unique()
```

```python
df['Regularizado']=df['Regularizado'].replace({'sim':'Sim','não':'Não','nao':'Não'})
```

```python
reg_counts = df['Regularizado'].value_counts().reset_index()
reg_counts.columns = ['Regularizado', 'Quantidade']
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=reg_counts, x='Regularizado', y='Quantidade', palette='Set2')
plt.title('Situação de Regularização dos Negócios')
plt.xlabel('Regularizado')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

🔹 4. Relação entre nota do Google e tipo de negócio

Objetivo: Avaliar quais segmentos têm mais problemas de reputação.

Valor: Entrar com soluções de padronização de processos, atendimento e segurança alimentar.

```python
#agrupar os dados por tipo de negócios e calcular a média da nota
nota_google = df.groupby('Tipo de negócio')['Nota google'].mean().reset_index()
nota_google = nota_google.sort_values(by='Nota google')
```

```python
#plotar gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=nota_google, x='Nota google', y='Tipo de negócio', palette='coolwarm')
plt.title('Média da Nota do google por Tipo de Negócio')
plt.xlabel('Nota Média')
plt.ylabel('Tipo de negócio')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

🔹 5. Presença digital (Instagram)

Objetivo: Identificar se os negócios estão posicionados online.

Valor: Negócios sem presença digital podem se beneficiar de soluções de branding, identidade, conteúdo e consultoria de imagem — excelente oportunidade para uma parceria com alguém da área de marketing, inclusive.

```python
#criar uma nova coluna binária indicando se o instagram foi informado ou não
df['Tem Instagram'] = df['Instagram'].notnull() & (df['Instagram'].str.strip() != '')
```

```python
#contar quantos negócios tem ou não tem instagram
instagram_counts = df['Tem Instagram'].value_counts()
```

```python
#plotar gráfico de pizza
plt.figure(figsize=(6,6))
plt.pie(instagram_counts, labels=['Com Instagram', 'Sem Instagram'], autopct='%1.1f%%', colors=['#6c9bcf','#f4a261'], startangle=140)
plt.title('Presença no Instagram entre os Negócios')
plt.axis('equal')
plt.show()
```

```python
#cruzamento com o tipo de negócio
presenca_por_tipo = df.groupby('Tipo de negócio')['Tem Instagram'].mean().reset_index()
presenca_por_tipo['Tem Instagram (%)'] = presenca_por_tipo['Tem Instagram'] * 100
presenca_por_tipo = presenca_por_tipo.sort_values(by='Tem Instagram (%)')

plt.figure(figsize=(10,6))
sns.barplot(data=presenca_por_tipo, x='Tem Instagram (%)', y='Tipo de negócio', palette='viridis')
plt.title('Porcentagem de Negócios com Instagram por Tipo')
plt.xlabel('% com Instagram')
plt.ylabel('Tipo de Negócio')
plt.tight_layout()
plt.show()
```
